## Multinomial Naive Bayes (Number of Ingredients by Cuisine Types)

This notebook requires:
* trainEngineered.csv

```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
finalDF = pd.read_csv('trainEngineered.csv')
finalDF
```

### Split into Train and Test

```python
X_train, X_test, y_train, y_test = train_test_split(finalDF.drop(['cuisine'], axis = 1), 
                                                    finalDF['cuisine'], 
                                                    train_size = 0.8, 
                                                    random_state = 5)
```

## Multinomial Naive Bayes

```python
nb_model = MultinomialNB()
```

```python
nb_model.fit(X_train, y_train)
```

```python
print("Score: ", nb_model.score(X_train, y_train))
```

```python
print(f'Accuracy: {nb_model.score(X_test, y_test)}')
```

We can see that Multinomial Naive Bayes performs even worse than that of random forest, gradient boosting, and support vector machine.
